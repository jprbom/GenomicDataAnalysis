```python
def naive(p, t):
    occurrences = []
    for i in range(len(t) - len(p) + 1):  # loop over alignments
        match = True
        for j in range(len(p)):  # loop over characters
            if t[i+j] != p[j]:  # compare characters
                match = False
                break
        if match:
            occurrences.append(i)  # all chars matched; record
    return occurrences
```

```python
def reverseComplement(s):
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A', 'N': 'N'}
    t = ''
    for base in s:
        t = complement[base] + t
    return t
```

```python
def readGenome(filename):
    genome = ''
    with open(filename, 'r') as f:
        for line in f:
            # ignore header line with genome information
            if not line[0] == '>':
                genome += line.rstrip()
    return genome
```

```python
def readFastq(filename):
    sequences = []
    qualities = []
    with open(filename) as fh:
        while True:
            fh.readline()  # skip name line
            seq = fh.readline().rstrip()  # read base sequence
            fh.readline()  # skip placeholder line
            qual = fh.readline().rstrip() # base quality line
            if len(seq) == 0:
                break
            sequences.append(seq)
            qualities.append(qual)
    return sequences, qualities
```

## Implement naive_with_rc

### Subtask:
Implement the `naive_with_rc` function. This function will take a pattern P and a text T. It will use the existing `naive` function to find occurrences of P in T and also find occurrences of the reverse complement of P in T using the `reverseComplement` function. It will return a sorted list of unique start offsets, ensuring that if P and its reverse complement are identical, any given match offset is reported only once.


**Reasoning**:
I need to implement the `naive_with_rc` function as described in the instructions, which involves using the previously defined `naive` and `reverseComplement` functions to find occurrences of a pattern and its reverse complement, then combining, deduplicating, and sorting the results.



```python
def naive_with_rc(p, t):
    occurrences_p = naive(p, t)
    p_rc = reverseComplement(p)
    occurrences_p_rc = naive(p_rc, t)
    combined_occurrences = occurrences_p + occurrences_p_rc
    unique_occurrences = set(combined_occurrences)
    sorted_occurrences = sorted(list(unique_occurrences))
    return sorted_occurrences

print("naive_with_rc function implemented.")
```

Implement the `naive_2mm` function to find occurrences of a pattern in a text allowing up to 2 mismatches. Then, use this function to search for the pattern 'TTCAAGCC' in the `lambda_genome` with a maximum of 2 mismatches, and finally, report the total number of occurrences found.

## Implement naive_2mm function

### Subtask:
Implement the `naive_2mm` function. This function will take a pattern P, a text T, and allow up to 2 mismatches. It should return a list of all start offsets where P (with up to 2 mismatches) occurs in T. It should *not* consider the reverse complement of P.


**Reasoning**:
I need to implement the `naive_2mm` function as described, following the detailed instructions provided in the subtask.



```python
def naive_2mm(p, t, max_mismatches=2):
    occurrences = []
    for i in range(len(t) - len(p) + 1):  # loop over alignments
        mismatches = 0
        for j in range(len(p)):  # loop over characters
            if t[i+j] != p[j]:  # compare characters
                mismatches += 1
            if mismatches > max_mismatches:
                break  # exceed max mismatches, no need to check further
        if mismatches <= max_mismatches:
            occurrences.append(i)  # all chars matched with <= max_mismatches
    return occurrences

print("naive_2mm function implemented.")
```

## Analyze quality scores per cycle

### Subtask:
Iterate through all the reads and their quality scores. For each sequencing cycle (position), calculate the average quality score across all reads. Convert Phred ASCII quality scores to numerical values (e.g., using `ord(char) - 33`).


**Reasoning**:
I need to initialize data structures and then iterate through the `qualities` list, convert ASCII quality scores to numerical Phred scores, and accumulate sums and counts for each cycle to calculate average quality scores.



```python
read_length = len(qualities[0])
total_quality_per_cycle = [0] * read_length
read_count_per_cycle = [0] * read_length

for qual_string in qualities:
    for i, char in enumerate(qual_string):
        phred_score = ord(char) - 33
        total_quality_per_cycle[i] += phred_score
        read_count_per_cycle[i] += 1

average_qualities_per_cycle = []
for i in range(read_length):
    if read_count_per_cycle[i] > 0:
        average_qualities_per_cycle.append(total_quality_per_cycle[i] / read_count_per_cycle[i])
    else:
        average_qualities_per_cycle.append(0) # Should not happen if all reads have the same length

print(f"Average quality scores calculated for {len(average_qualities_per_cycle)} cycles.")
print(f"First 10 average quality scores: {average_qualities_per_cycle[:10]}")
```
